```python
import numpy as np
import random
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-50, 50, 5000)
def func2(_x): return (_x**2)*(np.sin(0.5*_x)**2+1)
plt.plot(x, func2(x))
plt.show()

```

```python
def grad(_x): return (2*_x)*(np.sin(0.5*_x)*np.cos(0.5*_x))
```

```python
x= 20
alpha= 0.0001
n_iter=1000

for i in range(n_iter):
    print(x, func2(x))
    x = x - alpha*grad(x)
    x-=alpha*grad(x)
 
```

```python
from scipy.optimize import minimize
minimize(func2, np.random.uniform(-10, 10), jac = grad)
```

```python
class optimizer:
    def __init__(self, alpha, n_iter):
        self.alpha=alpha
        self.n_iter=n_iter
    def fit(self, func, grad, x0):
        for i in range(self.n_iter):
            x0-=self.alpha*grad(x0)
        return x0
opt1=optimizer(alpha, 75)
x=opt1.fit(func2, grad, np.random.uniform(-50, 50))
print(x, func2(x))
```
