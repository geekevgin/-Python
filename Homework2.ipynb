```python
#task2
from pylab import rcParams

rcParams["figure.figsize"] = 5.5, 7
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
print(f)
```

```python
plt.plot(t, f, color = 'green')
plt.title("График f(t)",fontsize=16)
plt.xlabel("Значения t", fontsize=14)
plt.ylabel("Значения f", fontsize=16)
plt.axis([-0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
#task3
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
print(y1)
```

```python
y2 = 2 * x + 0.5
y2
```

```python
y3 = -3 * x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
#Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotli



fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

ax1.set_title('График y1')
ax2.set_title('График y2')
ax3.set_title('График y3')
ax4.set_title('График y4')
ax1.set_xlim([-5, 5])
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

```python
class_0 = data.loc[data['Class'] == 0, ['V1']]
class_1 = data.loc[data['Class'] == 1, ['V1']]

```

```python
plt.hist(class_0['V1'], bins=20, density=True, alpha=0.5, label='Class_0', color='grey')
plt.hist(class_1['V1'], bins=20, density=True, alpha=0.5, label='Class_1', color='red')
plt.xlabel("V", fontsize=14)
plt.legend()
plt.show()
```
