```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
data = boston['data']
```

```python
feature_names = boston['feature_names']
```

```python
X = pd.DataFrame(data, columns = feature_names)
X.head()
```

```python
target = boston["target"]
```

```python
y = pd.DataFrame(target, columns= ['price'])
y.head()

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
check_test_lr = pd.DataFrame({
    "y_test": y_test["price"], 
    "y_pred_lr": y_pred_lr.flatten()})

check_test_lr.head()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred_lr)
```

```python
from  sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth= 12, random_state= 42)

```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred_model = model.predict(X_test)
check_test_model = pd.DataFrame({
    "y_test": y_test["price"], 
    "y_pred_model": y_pred_model.flatten()})

check_test_model.head()
```

```python
r2_score(y_test, y_pred_model)
```

```python
# Модель RandomForest работет лучше, так как выдает более точный результат. 
```

```python
model.feature_importances_
```

```python
feature_importance = pd.DataFrame({'name':X.columns, 
                                   'feature_importance':model.feature_importances_}, 
                                  columns=['feature_importance', 'name'])
feature_importance.head(10)
```

```python
feature_importance.nlargest(2, 'feature_importance')
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df['Class'].value_counts(normalize = True)
```

```python
df.info()
```

```python
pd.options.display.max_columns=100
```

```python
df.head(10)
```

```python
X = df.drop('Class', axis = 1)
X.head()
```

```python
y = df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 100, stratify= y)
```

```python
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
model = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3,
)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python

model = RandomForestClassifier(n_estimators= 15, max_depth= 6, max_features=3)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_test)
y_pred
```

```python
y_pred_proba = y_pred[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
