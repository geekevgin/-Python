```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import recall_score, precision_score, roc_auc_score, accuracy_score, f1_score
from sklearn import preprocessing
```

```python
data = pd.read_csv('MIT-BIH Arrhythmia Database.csv') #аритмии, данные основных зубцов и интервалов экг.
data.head()
```

```python
data.shape
```

```python
data['type'].value_counts() # виды аритмии и норма
```

```python
def evaluate_results(y_test, y_predict):
    print('Classification results:')
    f1 = f1_score(y_test, y_predict)
    print("f1: %.2f%%" % (f1 * 100.0)) 
    roc = roc_auc_score(y_test, y_predict)
    print("roc: %.2f%%" % (roc * 100.0)) 
    rec = recall_score(y_test, y_predict, average='binary')
    print("recall: %.2f%%" % (rec * 100.0)) 
    prc = precision_score(y_test, y_predict, average='binary')
    print("precision: %.2f%%" % (prc * 100.0)) 
    
    return f1, roc, rec, prc


res = [] # запишем общие результаты обучения
```

```python
data['type'] = data['type'].map({'VEB': 1, 'SVEB':1, 'F':1, 'Q':1, 'N': 0}) #сделаем бинаризацию, всю аномалию в 1, норму в 0
```

```python
x_data = data.iloc[:, 2:]
y_data = data[['type']]
```

```python
y_data.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data,test_size = 0.2, random_state=42) #соборьем на тест и обучение
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_train_scaled = min_max_scaler.fit_transform(X_train)
X_test_scaled = min_max_scaler.transform(X_test)
#нормализуем данные 
```

```python
model = xgb.XGBClassifier()

model.fit(X_train, y_train)
y_predict = model.predict(X_test)
```

```python
res.append(['pure target', *evaluate_results(y_test, y_predict)])
```

```python
#модель PU learning
#собираем рабочие позитивные samples
mod_data = data.copy() #дублируем 
#get the indices of the positives samples
pos_ind = np.where(mod_data['type'].values == 1)[0] #возьмем часть позитивов
#shuffle them
np.random.shuffle(pos_ind)
# leave just 25% of the positives marked
pos_sample_len = int(np.ceil(0.25 * len(pos_ind)))
print(f'Using {pos_sample_len}/{len(pos_ind)} as positives and unlabeling the rest')
pos_sample = pos_ind[:pos_sample_len]
```

```python
#делаем разметку
mod_data['class_test'] = -1
mod_data.loc[pos_sample, 'class_test'] = 1
print('target variable:\n', mod_data['class_test'].value_counts())

#без разметки 98037 примеров
```

```python
mod_data.head(3)
```

```python
x_data = mod_data.drop(columns=['type', 'class_test']).values # just the X 
y_labeled = mod_data['class_test'].values # new class (just the P & U)
y_positive = mod_data['type'].values # original class
```

```python
mod_data = mod_data.sample(frac=1)
#отделяем в данные в позитивы(1) и неразмеченные(-1)
data_N = mod_data[mod_data['class_test'] == -1] 
data_P = mod_data[mod_data['class_test'] == 1] 

neg_sample = data_N[:data_P.shape[0]] #количество негативов, равное кол-ву позитиков, для баланса 
sample_test = data_N[data_P.shape[0]:]
pos_sample = data_P.copy() 

print(neg_sample.shape, pos_sample.shape)
sample_train = pd.concat([neg_sample, pos_sample]).sample(frac=1)
```

```python
#классификатор строим. Он обучился на позитивных и случайных негативов
model = xgb.XGBClassifier()
sample_train.loc[sample_train['class_test']==-1, 'class_test'] = 0 # потому что XGBClassifier обучается на 0/1, а у нас -1/1

model.fit(sample_train.drop(columns=['type', 'class_test']).values, 
          sample_train['class_test'].values)
y_predict = model.predict(sample_test.drop(columns=['type', 'class_test']).values)
res.append(['random negative sampling', *evaluate_results(sample_test['type'].values, y_predict)])
```

```python
pd.DataFrame(res, columns=['method', 'f1', 'roc', 'recall', 'precision'])
#RNS стал хуже, отсев неверных результатов стал лучше, но точность нахождения верных стала ниже
```
