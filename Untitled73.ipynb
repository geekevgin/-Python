```python
not_mask = imgs_set.difference(masks_set)

not_mask = [f'image{i}.jpg' for i in not_mask]

df_lips_images  = df_lips_images .loc[~df_lips_images ['filename'].isin(not_mask)]
df_lips_images .reset_index(drop=True, inplace=True)
```

```python
SIZE = 256
class Lips_Segmentation_Dataset(Dataset):
    
    def __init__(self, list_frame, preprocessing=None):
        self.data = list_frame

        self.data_len = len(self.data.index)
        
        self.preprocessing = preprocessing
        
    def __len__(self):
        return self.data_len
    
    def __getitem__(self, idx):
        data = self.data.iloc[idx]
        img_path = os.path.join(IMAGE_DATASET_PATH, data['filename'])
        mask_path = os.path.join(MASK_DATASET_PATH, data['mask'])
        img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_RGB2BGR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = np.array(img).astype('float')
        
        mask = cv2.cvtColor(cv2.imread(mask_path), cv2.COLOR_RGB2BGR)
        mask = cv2.resize(mask, (SIZE, SIZE))
        mask = np.array(mask).astype('float')
        mask = torch.as_tensor(mask)

        if self.preprocessing:
            img = self.preprocessing(img)
            img = torch.as_tensor(img)
            mask = self.preprocessing(mask)
            mask = torch.as_tensor(mask)
            
            
        else:
          img = torch.as_tensor(img) / 255.0
          mask = torch.as_tensor(mask) / 255.0
        
        
        img = img.permute(2,0,1)
        mask = mask.permute(2,0,1)

        
        return (img.float(), mask) 
```

```python
# создаем даталоудеры 
train_dataset = Lips_Segmentation_Dataset(train)

test_dataset = Lips_Segmentation_Dataset(test)

train_data_loader = DataLoader(
    train_dataset,
    batch_size=24,
    shuffle=True
)

test_data_loader = DataLoader(
    test_dataset,
    batch_size=8,
    shuffle=False
)
```

```python
# Метрики, функциz потерь, оптимизатор
criterion = smp.utils.losses.DiceLoss()
metrics = [smp.utils.metrics.IoU(),]

optimizer = torch.optim.Adam(params=model.parameters(), lr=0.0001)
```

```python
# Обучение модель
train_epoch = smp.utils.train.TrainEpoch(
    model, 
    loss=criterion, 
    metrics=metrics, 
    optimizer=optimizer,
    device=device,
    verbose=True,
)

valid_epoch = smp.utils.train.ValidEpoch(
    model, 
    loss=criterion, 
    metrics=metrics, 
    device=device,
    verbose=True,
)
```

```python
for i in range(0, 5):
    print(f'Epoch: {i + 1}')
    train_logs = train_epoch.run(train_data_loader)
    valid_logs = valid_epoch.run(test_data_loader)
    
```

```python
def get_orig(image):
    image = image.permute(1, 2, 0)
    image = image.numpy()
    image = np.clip(image, 0, 1)
    return image
```

```python
for i, data in enumerate(test_data_loader):
    images, labels = data
    images = images.to(device)
    labels = labels.to(device)
    model.eval()
    outputs = model(images)
    f, axarr = plt.subplots(1,3, figsize=(15, 6))

    for j in range(0, 4):
        axarr[0].imshow(outputs.squeeze().detach().cpu().numpy()[j,:,:])
        axarr[0].set_title('Guessed labels')
        axarr[1].imshow(labels.squeeze().detach().cpu().numpy()[j,:, :].transpose(1,2,0))
        axarr[1].set_title('Truth labels')

        original = get_orig(images[j].cpu())
        axarr[2].imshow(original)
        axarr[2].set_title('Original Images')
        plt.show()
    if i > 6:
        break
```
